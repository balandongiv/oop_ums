The line `import matplotlib.pyplot as plt` is a common Python import statement used when working with the Matplotlib library. 

It's a way to include Matplotlib's `pyplot` module in our program.

```python
import matplotlib.pyplot as plt
```

```python
# Sample data
x = [1, 2, 3, 4, 5]

# Create a plot
plt.plot(x)

# Display the plot. In some environments, this command is not necessary.For example, in Jupyter Notebook, plots are displayed automatically after the cell that creates the plot is executed. You will notice, after this, I will omit the plt.show() command.
plt.show()
```

```python
# In this example, we will use the same list value for the x variable, and we introduce new y variable values.
y=[1,4,9,16,25]
plt.plot(x,y)
# plt.show()
```

In the previous example, we plotted a line graph by passing in only one list of numbers to `plt.plot()` but without the axis values. This is confusing because we don't know what the numbers on the x-axis and y-axis represent.

We can fix this by including the axis values as the first and second arguments in `plt.plot()`. The first argument is the list of numbers for the x-axis and the second argument is the list of numbers for the y-axis.

```python
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
# plt.show()
```

The plot() function has an optional third argument that specifies the appearance of the data points.
The default is b-, which is the blue solid line seen in the last two examples. The full list of styles can be found in the documentation for the plot() on the Matplotlib page


In the example below, we create a plot with blue circles for the data points. The third argument is 'bo', where 'b' stands for blue and 'o' stands for circle.

```python
plt.plot(x,y,'bo')
# plt.show()
```

In the example below, we create a plot with blue circles for the data points. The third argument is 'r+', where 'r' stands for red and '+' stands for start.

```python
plt.plot(x,y,'r+')
# plt.show()
```

You can quite easily alter the properties of the line with the plot() function. For example, you can change the size of the stars with the markersize argument.

```python
plt.plot(x,y,'r+',markersize=20)
```

or you can change the width of the line with the linewidth argument.

```python
plt.plot(x,y,linewidth=5)
```

## Altering tick labels

The tick labels are the numbers that you see on the x-axis and y-axis. You can manually alter the appearance of the tick labels with the `xticks()` and `yticks()` functions. The `xticks()` function alters the appearance of the tick labels on the x-axis and the `yticks()` function alters the appearance of the tick labels on the y-axis.

<font color="red">**!Note that the tick values have to be contained within a list object.**</font>


```python
plt.plot([1,2,3,4,5],[1,4,9,20,2],'g--',linewidth=5)
plt.xticks([0,5,10])
plt.yticks([0,25,50,60])
plt.show()
```

## The axis() function

The axis() function allows us to specify the range of the axis.

It is a convenient way to control the visible range of your data on the plot. 

It requires a list that contains the following

The function takes four arguments:

`plt.axis([xmin, xmax, ymin, ymax])`

Where:

`xmin`: `The minimum value for the x-axis`.
`xmax`: `The maximum value for the x-axis`.
`ymin`: `The minimum value for the y-axis`.
`ymax`: `The maximum value for the y-axis`.


```python
plt.plot([1,2,3,4,5],[1,4,9,20,2],'g--',linewidth=5)

# Here, we set the x-axis range from 0 to 6 and the y-axis range from 2 to 30
plt.axis([0,6,2,30])  # [xmin, xmax, ymin, ymax]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a range of x values
x = np.linspace(0, 2 * np.pi, 100)
# Define the cosine function with a shift and amplitude
A = 2  # Amplitude
B = 1  # Shift
y = A * np.cos(x - B)

# Find the maximum point
max_x = x[np.argmax(y)]
max_y = A * np.cos(max_x - B)

# Plot the cosine function
plt.plot(x, y)  

# Add labels and title
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.title(r'$y = 2\cos(x - 1)$')



# Add text at the maximum point
plt.text(max_x+0.5, max_y-0.2, f'Max: ({max_x:.2f}, {max_y:.2f})', fontsize=12, color='red')

# Show the plot
plt.grid(True)
plt.show()
```

## Annotations

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a range of x values
x = np.linspace(0, 2 * np.pi, 100)
# Define the cosine function with a shift and amplitude
A = 2  # Amplitude
B = 1  # Shift
y = A * np.cos(x - B)

# Find the maximum point
max_x = x[np.argmax(y)]
max_y = A * np.cos(max_x - B)

# Plot the cosine function
plt.plot(x, y)  # TeX equation expression

# Annotate the maximum point with an arrow and text
plt.annotate(
    f'Maximum: ({max_x:.2f}, {max_y:.2f})',  # Text to display
    xy=(max_x, max_y),  # Position of the point to annotate
    xytext=(max_x + 1.5, max_y - 1),  # Position of the text (adjusted to be below the maximum point and does not intersect with the line)
    arrowprops=dict(arrowstyle='->'),  # Arrow style
    fontsize=12,
)

# Add labels and title
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.title(r'$y = 2\cos(x - 1)$')


# Add a legend
# plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
