## Adding Titles and Labels

In the previous example, we plotted a graph by passing in only one list of numbers to `plt.plot()` but without the axis values. This is confusing because we don't know what the numbers on the x-axis and y-axis represent.

To make the plot more informative, we can add labels to the x-axis and y-axis. We can also add a title to the plot.

```python
# Sample data
xdata = rnd.randn(100) * 1000
ydata = rnd.randn(100) * 20 + 25

plt.scatter(xdata, ydata)

# Customize X and Y axis labels
plt.xlabel("Solar irradiance (W/m^2)")
plt.ylabel("Temperature (C)")
plt.title("Solar irradiance vs temperature in Malaysia")

# plt.show()
```

These labels provide a clear context for the data being plotted, which is related to solar renewable energy.

You can quite easily alter the properties of the line with the plot() function. For example, you can change the size of the stars with the markersize argument.

```python
import numpy as np

# Sample data
# Generate random data for solar irradiance and temperature
solar_irradiance = np.random.randn(100) * 1000
temperature = np.random.randn(100) * 20 + 25


# Create a scatter plot of solar irradiance vs temperature
plt.scatter(solar_irradiance, temperature, s=100, marker='x')

# Add labels and title
plt.xlabel("Solar irradiance (W/m^2)")
plt.ylabel("Temperature (C)")
plt.title("Solar irradiance vs temperature in Malaysia")
```

## Setting Axis Limits

We sometimes want to set the axis limits to a specific range  to zoom in on a plot or to zoom out on a plot.

Temporary
Zooming In or Out: Sometimes, your data might have outliers or extreme values that make it difficult to see the main trends or patterns in your plot. By setting custom axis limits, you can "zoom in" on a specific range of data points or "zoom out" to get a broader view.

Comparing Plots: When you have multiple plots or subplots on the same figure, setting consistent axis limits across them can make it easier to compare different datasets. This ensures that the scale is the same for all plots, facilitating a more accurate visual comparison.

Emphasizing Specific Data Ranges: You might want to emphasize a particular range of data in your plot. By setting axis limits, you can focus on a specific portion of the data to highlight trends, patterns, or changes.

Removing Irrelevant Data: Sometimes, data may extend beyond a meaningful or relevant range. Setting axis limits allows you to exclude irrelevant data points from the visualization, making the plot more informative and cleaner.

Customizing the Visualization: Axis limits can be used as a design choice to control the visual appearance of your plot. You can adjust limits to frame your data in a way that conveys the message you want to communicate effectively.

For example, we want the x-axis to range from 0 to 6 and the y-axis to range from 0 to 30. We can do this with the `plt.axis()` function.

The plot() function has an optional third argument that specifies the appearance of the data points.
The default is b-, which is the blue solid line seen in the last two examples. The full list of styles can be found in the documentation for the plot() on the Matplotlib page


In the example below, we create a plot with blue circles for the data points. The third argument is 'bo', where 'b' stands for blue and 'o' stands for circle.

The axis() function allows us to specify the range of the axis.

It is a convenient way to control the visible range of your data on the plot. 

It requires a list that contains the following

The function takes four arguments:

`plt.axis([xmin, xmax, ymin, ymax])`

Where:

`xmin`: `The minimum value for the x-axis`.
`xmax`: `The maximum value for the x-axis`.
`ymin`: `The minimum value for the y-axis`.
`ymax`: `The maximum value for the y-axis`.


```python
# Create a scatter plot of solar irradiance vs temperature
plt.scatter(solar_irradiance, temperature, s=100, marker='x')

# For clarity, we omit the x-axis and y-axis labels and title

# Here, we set the x-axis range from 0 to 6 and the y-axis range from 2 to 30
plt.axis([-2000, 1500,-10, 60])  # [xmin, xmax, ymin, ymax]

```

## Altering tick labels

The tick labels are the numbers that you see on the x-axis and y-axis. You can manually alter the appearance of the tick labels with the `xticks()` and `yticks()` functions. The `xticks()` function alters the appearance of the tick labels on the x-axis and the `yticks()` function alters the appearance of the tick labels on the y-axis.

<font color="red">**!Note that the tick values have to be contained within a list object.**</font>


```python
plt.scatter(solar_irradiance, temperature, s=100, marker='x')
# To improve clarity, we have omitted the setting of x-axis and y-axis labels, title, and axis limits.

# Customize the tick labels on the x-axis and y-axis
plt.xticks([-1500,5,1000])
plt.yticks([0,25,50,60])
plt.show()
```

## Add text at some specific location

```python
plt.scatter(solar_irradiance, temperature, s=100, marker='x')
# To improve clarity, we have omitted the setting of x-axis and y-axis labels, title, and axis limits.

# Say for example, there is irregularity in the data and we want to highlight that particular point.

irregularity_x,irregularity_y = 0, -10

# Add text at a specific position
plt.scatter(irregularity_x,irregularity_y, s=100, marker='o', color='red')
plt.text(irregularity_x,irregularity_y, 'Maximum', fontsize=12, color='red')
plt.show()
```

## Adding Annotations with Arrows

```python
# Create a scatter plot of solar irradiance vs temperature with cross markers
plt.scatter(solar_irradiance, temperature, s=100, marker='x')

# Irregularity point
irregularity_x, irregularity_y = 0, -10

# Add a red circle marker and label for the irregularity point
plt.scatter(irregularity_x, irregularity_y, s=100, marker='o', color='red')

# Annotate the irregularity point with an arrow
plt.annotate(
    'Arrow Pointing Here',
    xy=(irregularity_x, irregularity_y),
    xytext=(20, -20),  # Position of the text relative to the point
    textcoords='offset points',
    arrowprops=dict(arrowstyle='->', color='blue', linewidth=2),
    fontsize=12,
    color='blue'
)

# To improve clarity, you can add labels, a title, and axis limits here if needed.

# Show the plot
plt.show()
```

## Adding Legends

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for Peninsular Malaysia
solar_irradiance_peninsular = np.random.randn(50) * 1000
temperature_peninsular = np.random.randn(50) * 20 + 30  # Slightly higher temperature

# Sample data for Borneo
solar_irradiance_borneo = np.random.randn(50) * 800  # Lower irradiance
temperature_borneo = np.random.randn(50) * 20 + 25

# Create a scatter plot for Peninsular Malaysia
plt.scatter(solar_irradiance_peninsular, temperature_peninsular, label='Peninsular Malaysia', s=50, marker='o')

# Create a scatter plot for Borneo
plt.scatter(solar_irradiance_borneo, temperature_borneo, label='Borneo', s=50, marker='x')

# Add labels and title
plt.xlabel("Solar irradiance (W/m^2)")
plt.ylabel("Temperature (C)")
plt.title("Comparison of Solar Irradiance and Temperature between Peninsular Malaysia and Borneo")

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)  # Add a grid for better visualization
plt.show()
```

## Multiple Plots

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the first subplot
years = [2016, 2017, 2018, 2019, 2020]
solar_capacity = [40, 45, 50, 55, 60]  # in megawatts

# Sample data for the second subplot
# Generate random data for solar irradiance and temperature
solar_irradiance = np.random.randn(100) * 1000
temperature = np.random.randn(100) * 20 + 25

# Create subplots with 1 row and 2 columns
plt.figure(figsize=(14, 6))

# First subplot
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.scatter(years, solar_capacity, label='Solar Capacity (MW)', color='b', marker='o')
plt.xlabel('Year')
plt.ylabel('Solar Capacity (MW)')
plt.title('Solar Energy Capacity in Malaysia (2010-2020)')
plt.grid(True)
plt.legend()

# Second subplot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.scatter(solar_irradiance, temperature, s=100, marker='x')
plt.xlabel('Solar Irradiance')
plt.ylabel('Temperature (C)')
plt.title('Solar Irradiance vs. Temperature')
plt.grid(True)

# Adjust spacing between subplots
plt.tight_layout()

# Show the subplots
plt.show()
```

## Saving Plots

```python
# Create a plot
plt.plot([1, 2, 3, 4, 5])

# Save the plot as an image (e.g., PNG)
plt.savefig('subplots.png')
```

Other specific settings can be used to control the resolution of the image, the size of the image, and the background color. For example, we can set the background color to white with the `facecolor` argument.

```python
# Create a plot
plt.plot([1, 2, 3, 4, 5])

# Save the plot as an image (e.g., PNG)
plt.savefig('subplots.png',         # File name of the plot image
            facecolor='white',      # Background color of the image
            dpi=300,                # Resolution of the image
            bbox_inches='tight',     # Crop the extra white space
            transparent=True)      # Set to True if you want the background to be transparent
```

## Using SeaBorn

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Sample data for the first subplot
years = [2016, 2017, 2018, 2019, 2020]
solar_capacity = [40, 45, 50, 55, 60]  # in megawatts

# Sample data for the second subplot
# Generate random data for solar irradiance and temperature
solar_irradiance = np.random.randn(100) * 1000
temperature = np.random.randn(100) * 20 + 25

# Set Seaborn style
sns.set(style="whitegrid")

# Create subplots with 1 row and 2 columns
plt.figure(figsize=(14, 6))

# First subplot
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
sns.scatterplot(x=years, y=solar_capacity, label='Solar Capacity (MW)', color='b', marker='o')
plt.xlabel('Year')
plt.ylabel('Solar Capacity (MW)')
plt.title('Solar Energy Capacity in Malaysia (2010-2020)')
plt.legend()

# Second subplot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
sns.scatterplot(x=solar_irradiance, y=temperature, s=100, marker='x')
plt.xlabel('Solar Irradiance')
plt.ylabel('Temperature (C)')
plt.title('Solar Irradiance vs. Temperature')

# Adjust spacing between subplots
plt.tight_layout()

# Show the subplots
plt.show()
```
